```python
import pandas as pd
```

```python
# Series.str.split(pat=None, n=-1, expand=False)
```

```python
#原始迁入数据
address = pd.read_excel('/Users/jackchen/Desktop/迁徙地图（市）.xlsx', index_col=False)
```

```python
address
```

```python
#方法一 split分割`市`
```

```python
address1 = address['名称'].str.split('市', expand=True)
address1
```

```python
address1.loc[:, 2] = address.iloc[:, 1] 
```

```python
address1
```

```python
#合并表格
address2 = address.join(address['名称'].str.split('市', expand=True))
address2
```

```python
#方法二 使用匿名函数lambda来将City列拆成两列。
```

```python
address['城市'] = address['名称'].map(lambda x:x. split('市')[0])
address['省份'] = address['名称'].map(lambda x:x. split('市')[-1])
```

```python
address
```

```python
df1 = address[['省份', '城市', '比例']]
```

```python
df1
```

```python
#迁出数据 操作同上
address = pd.read_excel('/Users/jackchen/Desktop/迁徙地图（市）.xlsx', sheet_name = '迁出（市）', index_col=False)
```

```python
address['城市'] = address['名称'].map(lambda x:x. split('市')[0])
address['省份'] = address['名称'].map(lambda x:x. split('市')[-1])
address
```

```python
df2 = address[['省份', '城市', '比例']]
```

```python
# 导出至同一Excel中的不同sheet，并命名。
writer = pd.ExcelWriter('6.3人口迁徙数据（市）.xlsx')
df1.to_excel(writer, sheet_name='6.3迁入（市）')
df2.to_excel(writer, sheet_name='6.3迁出（市）')
writer.save()
writer.close()
```
