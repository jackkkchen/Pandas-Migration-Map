```python
#方法一 split分割`市`

# split()函数语法：str.split(str="",num=string.count(str))[n]

# 参数说明：
# str:表示为分隔符，默认为空格，但是不能为空(’’)。若字符串中没有分隔符，则把整个字符串作为列表的一个元素
# num:表示分割次数。如果存在参数num，则仅分隔成 num+1 个子字符串，并且每一个子字符串可以赋给新的变量
# [n]:表示选取第n个分片
# expand: 指定expand=True生成DataFrame

# 注意：当使用空格作为分隔符时，对于中间为空的项会自动忽略
```

```python
address1 = address['名称'].str.split('市', expand=True)
address1
```

```python
address1.loc[:, 2] = address.iloc[:, 1] 
```

```python
address1
```

```python
#合并表格
address2 = address.join(address['名称'].str.split('市', expand=True))
address2
```

```python
#方法二 使用匿名函数lambda来将City列拆成两列。
```

```python
address['城市'] = address['名称'].map(lambda x:x. split('市')[0])
address['省份'] = address['名称'].map(lambda x:x. split('市')[-1])
```

```python
address
```

```python
df1 = address[['省份', '城市', '比例']]
```

```python
df1
```

```python
#迁出数据 操作同上
address = pd.read_excel('/Users/jackchen/Desktop/迁徙地图（市）.xlsx', sheet_name = '迁出（市）', index_col=False)
```
